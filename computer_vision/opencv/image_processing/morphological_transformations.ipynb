# Morphological Transformations

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img_path = os.path.join('..', '..', 'pics', 'j.png')
img = cv.imread(img_path, cv.IMREAD_GRAYSCALE)
plt.imshow(img, 'gray')
```

## Erosion
The thickness or size of the foreground object decreases or simply white region decreases in the image. It is useful for removing small white noises (as we have seen in colorspace chapter), detach two connected objects etc.

```python
kernel = np.ones((5,5),np.uint8)
erosion = cv.erode(img,kernel,iterations = 1)
plt.imshow(erosion, 'gray')
```

## Dilation
it increases the white region in the image or size of foreground object increases. Normally, in cases like noise removal, erosion is followed by dilation. Because, erosion removes white noises, but it also shrinks our object.

```python
dilation = cv.dilate(img,kernel,iterations = 1)
plt.imshow(dilation, 'gray')
```

## Opening
Opening is just another name of erosion followed by dilation. It is useful in removing noise, as we explained above.

```python
opening = cv.morphologyEx(img, cv.MORPH_OPEN, kernel)
plt.imshow(opening, 'gray')
```

## Closing
Closing is reverse of Opening, Dilation followed by Erosion. It is useful in closing small holes inside the foreground objects, or small black points on the object.

```python
closing = cv.morphologyEx(img, cv.MORPH_CLOSE, kernel)
plt.imshow(dilation, 'gray')
```

## Morphological Gradient
It is the difference between dilation and erosion of an image

```python
gradient = cv.morphologyEx(img, cv.MORPH_GRADIENT, kernel)
plt.imshow(gradient,'gray')
```

## Top Hat
It is the difference between input image and Opening of the image. Below example is done for a 9x9 kernel.

```python
kernel = np.ones((8,8),np.uint8)
tophat = cv.morphologyEx(img, cv.MORPH_TOPHAT, kernel)
plt.imshow(tophat,'gray')
```

## Black Hat
It is the difference between the closing of the input image and input image

```python
blackhat = cv.morphologyEx(img, cv.MORPH_BLACKHAT, kernel)
plt.imshow(blackhat,'gray')
```
