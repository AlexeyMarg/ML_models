# Image inpainting

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
img_path = os.path.join('..', '..', 'pics', 'pic.jpg')
img = cv.imread(img_path)
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
cv.rectangle(img,pt1=(20,20),pt2=(100,100),color=(255, 0, 0),thickness=5 )
plt.imshow(img)
```

```python
mask = np.zeros(shape=img.shape, dtype=np.uint8)
mask = cv.cvtColor(mask, cv.COLOR_RGB2GRAY)
cv.rectangle(mask, pt1=(20,20), pt2=(100,100),color=(255, 255, 255), thickness=5)
plt.imshow(mask, cmap='gray')
```

```python
dst = cv.inpaint(img,mask,3,cv.INPAINT_TELEA)
plt.imshow(dst)
```
